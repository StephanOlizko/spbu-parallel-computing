```python
# Построй график по csv файлу results/1.txt

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv('results/1.txt', sep=',') # Размер вектора, Кол-во процессов, Время последовательно, Время параллельно, Min, Max

data_8 = data[data['Кол-во процессов'] == 8]
data_16 = data[data['Кол-во процессов'] == 16]
data_32 = data[data['Кол-во процессов'] == 32]

fig, ax = plt.subplots()
ax.plot(data_8['Размер вектора'], data_8['Время последовательно'], label='Последовательно')
ax.plot(data_8['Размер вектора'], data_8['Время параллельно'], label='Параллельно')
ax.set_xlabel('Размер вектора')
ax.set_ylabel('Время')
ax.legend()
ax.set_yscale('log')
ax.set_xscale('log')
plt.show()

fig, ax = plt.subplots()
ax.plot(data_16['Размер вектора'], data_16['Время последовательно'], label='Последовательно')
ax.plot(data_16['Размер вектора'], data_16['Время параллельно'], label='Параллельно')
ax.set_xlabel('Размер вектора')
ax.set_ylabel('Время')
ax.legend()
ax.set_yscale('log')
ax.set_xscale('log')
plt.show()

fig, ax = plt.subplots()
ax.plot(data_32['Размер вектора'], data_32['Время последовательно'], label='Последовательно')
ax.plot(data_32['Размер вектора'], data_32['Время параллельно'], label='Параллельно')
ax.set_xlabel('Размер вектора')
ax.set_ylabel('Время')
ax.legend()
ax.set_yscale('log')
ax.set_xscale('log')
plt.show()


# Ускорение
fig, ax = plt.subplots()
ax.plot(data_8['Размер вектора'], data_8['Время последовательно'] / data_8['Время параллельно'], label='8 процессов')
ax.plot(data_16['Размер вектора'], data_16['Время последовательно'] / data_16['Время параллельно'], label='16 процессов')
ax.plot(data_32['Размер вектора'], data_32['Время последовательно'] / data_32['Время параллельно'], label='32 процесса')
ax.set_xlabel('Размер вектора')
ax.set_ylabel('Ускорение')
ax.set_xscale('log')
ax.legend()
plt.show()

```
