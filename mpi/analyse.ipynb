```python
#    csv  results/1.txt

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv('results/1.txt', sep=',') #  , - ,  ,  , Min, Max

data_8 = data[data['- '] == 8]
data_16 = data[data['- '] == 16]
data_32 = data[data['- '] == 32]

fig, ax = plt.subplots()
ax.plot(data_8[' '], data_8[' '], label='')
ax.plot(data_8[' '], data_8[' '], label='')
ax.set_xlabel(' ')
ax.set_ylabel('')
ax.legend()
ax.set_yscale('log')
ax.set_xscale('log')
plt.show()

fig, ax = plt.subplots()
ax.plot(data_16[' '], data_16[' '], label='')
ax.plot(data_16[' '], data_16[' '], label='')
ax.set_xlabel(' ')
ax.set_ylabel('')
ax.legend()
ax.set_yscale('log')
ax.set_xscale('log')
plt.show()

fig, ax = plt.subplots()
ax.plot(data_32[' '], data_32[' '], label='')
ax.plot(data_32[' '], data_32[' '], label='')
ax.set_xlabel(' ')
ax.set_ylabel('')
ax.legend()
ax.set_yscale('log')
ax.set_xscale('log')
plt.show()


# 
fig, ax = plt.subplots()
ax.plot(data_8[' '], data_8[' '] / data_8[' '], label='8 ')
ax.plot(data_16[' '], data_16[' '] / data_16[' '], label='16 ')
ax.plot(data_32[' '], data_32[' '] / data_32[' '], label='32 ')
ax.set_xlabel(' ')
ax.set_ylabel('')
ax.set_xscale('log')
ax.legend()
plt.show()

```

```python
#     csv  results/9_8.txt

data = pd.read_csv('results/9_8.txt', sep=',') # Operation,Type,Time (seconds)

#         
data_pair = data[data['Type'] == 'pairwise']
data_mpi = data[data['Type'] == 'MPI']

fig, ax = plt.subplots()
data_pair.plot(kind='bar', x='Operation', y='Time (seconds)', ax=ax)
ax.set_yscale('log')
ax.set_xlabel("Pairwise")
plt.show()

fig, ax = plt.subplots()
data_mpi.plot(kind='bar', x='Operation', y='Time (seconds)', ax=ax)
ax.set_yscale('log')
ax.set_xlabel("MPI")
plt.show()



data = pd.read_csv('results/9_16.txt', sep=',') # Operation,Time,Method

#         
data_pair = data[data['Method'] == 'Pairwise']
data_mpi = data[data['Method'] == 'MPI']

fig, ax = plt.subplots()
data_pair.plot(kind='bar', x='Operation', y='Time', ax=ax)
ax.set_yscale('log')
ax.set_xlabel("Pairwise")
plt.show()

fig, ax = plt.subplots()
data_mpi.plot(kind='bar', x='Operation', y='Time', ax=ax)
ax.set_yscale('log')
ax.set_xlabel("MPI")
plt.show()


data = pd.read_csv('results/9_32.txt', sep=',') # operation,time,method

#         
data_pair = data[data['method'] == 'Pairwise']
data_mpi = data[data['method'] == 'MPI']

fig, ax = plt.subplots()
data_pair.plot(kind='bar', x='operation', y='time', ax=ax)
ax.set_yscale('log')
ax.set_xlabel("Pairwise")
plt.show()

fig, ax = plt.subplots()
data_mpi.plot(kind='bar', x='operation', y='time', ax=ax)
ax.set_yscale('log')
ax.set_xlabel("MPI")
plt.show()






```
